### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
hygiene_score_20_query = {'scores.Hygiene': {'$eq': 20}}
hygiene_score_20_results = establishments.find(hygiene_score_20_query)

# Use count_documents to display the number of documents in the result
num_of_docs = establishments.count_documents(hygiene_score_20_query)
print(f'Number of establishments with a hygiene score of 20: {num_of_docs}\n')

# Display the first document in the results using pprint
print('The first document in the results showing a hygiene score of 20:\n')
pprint(hygiene_score_20_results[0])
```

```python
# Convert the result to a Pandas DataFrame
hygiene_score_20_results_df = pd.DataFrame(hygiene_score_20_results)

# Display the number of rows in the DataFrame
print(f'The number of rows in hygiene20_results_df: {len(hygiene_score_20_results_df)}')

# Display the first 10 rows of the DataFrame
hygiene_score_20_results_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
London_loc_auth_query =  {'LocalAuthorityName': {'$regex': 'London', "$options": "i"}, 'RatingValue': {'$gte': '4'}}
London_loc_auth_results = establishments.find(London_loc_auth_query)

# Use count_documents to display the number of documents in the result
number_of_docs = establishments.count_documents(London_loc_auth_query)
print(f'The number of documents with London as the Local Authority and with a RatingValue greater than or equal to 4: {number_of_docs}\n')

# Display the first document in the results using pprint
print('The first document in the results showing London as the Local Authority and with a RatingValue greater than or equal to 4:\n')
pprint(London_loc_auth_results[0])
```

```python
# Convert the result to a Pandas DataFrame
London_loc_auth_results_df = pd.DataFrame(London_loc_auth_results)

# Display the number of rows in the DataFrame
print(f'The number of rows in London_loc_auth_results_df: {len(London_loc_auth_results_df)}')

# Display the first 10 rows of the DataFrame
London_loc_auth_results_df.head(10)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
match_query = {'$match': {'scores.Hygiene': {'$eq':0}}}

# 2. Groups the matches by Local Authority
group_query = {'$group': {'_id':'$LocalAuthorityName', 'count':{'$sum': 1 }}}

# 3. Sorts the matches from highest to lowest
sort_values = {'$sort': {'count':-1}}

# Creating pipeline
pipeline = [match_query, group_query, sort_values]
pipeline_results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
numb_of_docs = establishments.count_documents({'scores.Hygiene': {'$eq':0}})
print(f"The number of documents in the pipeline results: {numb_of_docs}\n")

# Print the first 10 results
pprint(pipeline_results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
pipeline_results_df = pd.DataFrame(pipeline_results)

# Display the number of rows in the DataFrame
print(f'The number of rows in pipeline_results_df: {len(pipeline_results_df)}')

# Display the first 10 rows of the DataFrame
pipeline_results_df.head(10)
```
